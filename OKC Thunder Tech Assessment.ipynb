```python
import pandas as pd
import numpy as np
```

```python
allShots = pd.read_csv('shots_data.csv')
allShots
```

```python
#cretaing a column called "Distance" which is calculated by using Pytharean Theorem (r = sqrt(x^2 + y^2))
#This new column will represent the Distance from the hoop
allShots['Distance'] = np.sqrt((allShots['x']**2) + (allShots['y']**2))
allShots
```

### Separating data into the two teams (Team A & Team B)

```python
teamA = allShots[allShots['team'] == "Team A"]
teamA
```

```python
teamB = allShots[allShots['team'] == "Team B"]
teamB
```

# Finding Shot Distribution & eFG% for Team A
### Beginning by separating the shots across the court into a top and bottom portion. The top portion is the semicircular area and the bottom portion is the rectangular area. Both are separated by the line y = 7.8 ft.

```python
#separating the data into the bottom rectangular portion below the separation of arc and "corner" (where y <= 7.8 ft)
belowCornerA = teamA[teamA['y'] <= 7.8]
belowCornerA
```

```python
#separating the data into the top semicircular area above the "corner"
aboveCornerA = teamA[teamA['y'] > 7.8]
aboveCornerA
```

```python
#Creating a new table that contains every non corner 3 pointer (NC3)
#A non-corner 3 pointer is classified as a shot outside the arc (which is more than 23.75 ft away from the rim)
#23.75 ft shots are not counted because if a player is on the 3 point line, it is not considered a 3 pointer

NC3A = aboveCornerA[aboveCornerA['Distance'] > 23.75]
NC3A
```

```python
#Creating a new table that holds all 2 point shots that team A took
#I begin by taking all shots from above the corner (the upper semicircular area) where the shot was taken <= 23.75 ft away
A2PT = aboveCornerA[aboveCornerA['Distance'] <= 23.75]
#Here, I find the rest of the 2 point shots that were taken in the bottom half (rectangular area)
entry = belowCornerA.loc[belowCornerA['Distance'] <= 22.0]
#Finally, I add these rows to the team A 2pt shots taken table
A2PT = pd.concat([A2PT, entry])
A2PT
```

```python
#Creating the last table which contains all Corner 3 point shots (which are considered to be 22+ ft away)
C3A = belowCornerA[belowCornerA['Distance'] > 22.0]
C3A
```

### Finding shot distribution for Team A

```python
A2PTdist = "{:.3%}".format(A2PT.shape[0] / teamA.shape[0])
NC3Adist = "{:.3%}".format(NC3A.shape[0] / teamA.shape[0])
C3Adist = "{:.3%}".format(C3A.shape[0] / teamA.shape[0])
print("The shot distrution for Team A is as follows:")
print("\n2PT = " + A2PTdist)
print("\nNC3 = " + NC3Adist)
print("\nC3 = " + C3Adist)
```

### Finding eFG% for Team A in each shot zone (2pt, NC3, C3)

```python
#finding corner 3 eFG%
C3fgmA = len(C3A.loc[C3A['fgmade'] == 1])

C3fgaA = C3A.shape[0]

C3eFGA = "{:.3%}".format((C3fgmA + (0.5*C3fgmA))/C3fgaA)
print("Team A's Corner 3 eFG% (or effective field goal percentage) = " + C3eFGA)
```

```python
#finding non-corner 3 eFG%
NC3fgmA = len(NC3A.loc[NC3A['fgmade'] == 1])

NC3fgaA = NC3A.shape[0]

NC3eFGA = "{:.3%}".format((NC3fgmA + (0.5*NC3fgmA))/NC3fgaA)
print("Team A's Non-Corner 3 eFG% (or effective field goal percentage) = " + NC3eFGA)
```

```python
#finding 2 point eFG%
twoPTfgmA = len(A2PT.loc[A2PT['fgmade'] == 1])

twoPTfgaA = A2PT.shape[0]

# there are no 3 pointers made in the 2 point zone, so 3PM = 0.
threePM = 0

twoPTeFGA = "{:.3%}".format((twoPTfgmA + (0.5*threePM))/twoPTfgaA)
print("Team A's 2-point eFG% (or effective field goal percentage) = " + twoPTeFGA)
```
